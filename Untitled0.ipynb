#primeiro web scrapping
- usando Python para buscar preços de produtos no mercado livre

```python
import requests
from bs4 import BeautifulSoup
```

```python
url_site = 'https://lista.mercadolivre.com.br/'
```

```python
produto = input('Digite o nome do produto')
print(produto)
```

```python
produto_tratado = produto.replace(" ","-")
print(produto_tratado)
```

```python
link_produto = url_site + produto_tratado
print(link_produto)
```

```python
#ver se o site aceita requisiçao (coteudo do site - pagina em texto HTML)
response =  requests.get(link_produto)
```

```python
print(response)
```

```python
BeautifulSoup,
#html.parse é o analisador d HTML que iremos utilizar para analisar o conteudo do site
site = BeautifulSoup (response.text, 'html.parser')
```

```python
print(site)
```

```python

#find_all()= metodo da classe Beuatifulsoup que permite buscar todas as ocorrencias uma tag HTML especifica
dados = site.find_all("div", attrs = {"class" : "ui-search-result__content-wrapper shops__result-content-wrapper"})
```

```python
print (dados)
```

```python
nomes = []
preco = []

for x in dados:
  #nome do produto
  nome_produto = x.find_all("h2")[0].text

  #preço
  preco_produto = x.find_all("span", class_="andes-money-amount__fraction")[0].text

  nomes.append(nome_produto)
  preco.append(preco_produto)
```

```python
len(nomes)
```

```python
import pandas as pd
dados_coletados = {
    "nome": nomes,
    "preco": preco
    }
```

```python
pd.DataFrame(dados_coletados)
```
